```python
import tiktoken
```

```python
tokenizer = tiktoken.get_encoding("gpt2")
```

```python
with open("./docs/gospels.txt", "r", encoding="utf-8") as f:
    raw_text = f.read()

enc_text = tokenizer.encode(raw_text)
print(len(enc_text))
```

```python
token_sample = enc_text[:10]
print(token_sample)
```

```python
words = tokenizer.decode(token_sample)
print(words)
```
